#### Find faulty sections by plotting the data.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))
plt.plot(df_bikes['Start DateTime'], df_bikes['Fahrradbruecke total'], label='Fahrradbruecke')
plt.plot(df_bikes['Start DateTime'], df_bikes['Temperatur (°C)'], label='Temperatur')
plt.xlabel('Time')
plt.ylabel('Amount of cyclists')
plt.legend()
plt.title('Time Series Plot')
plt.show()

df_bikes.plot()
plt.show()

### Show daily aggregation

plt.figure(figsize=(14,6))

# Convert 'Start DateTime' to datetime if it's not already
df_bikes['Start DateTime'] = pd.to_datetime(df_bikes['Start DateTime'])

# Set 'Start DateTime' as the index
df = df_bikes.set_index('Start DateTime')

# Select only the columns to resample
df_to_resample = df[['Fahrradbruecke total']]

# Resample and aggregate by day
df_daily = df_to_resample.resample('D').sum()

plt.plot(df_daily.index, df_daily['Fahrradbruecke total'], label='Fahrradbruecke', linestyle='solid')

plt.xlabel('Time')
plt.ylabel('Cyclists Count')
plt.legend()
plt.title('Time Series Plot (Daily Aggregate)')
plt.show()

plt.figure(figsize=(14,6))

# Filter data for the specific day
df_day = df[df.index.date == pd.to_datetime('2021-01-15').date()]

plt.plot(df_day.index, df_day['Fahrradbruecke total'], label='Fahrradbruecke', linestyle='solid')

plt.xlabel('Time')
plt.ylabel('Count')
plt.legend()
plt.title('Time Series Plot for 2021-01-15')
plt.show()
```

There is a known problem in 2021 with the bike counter. Find the exact timeframe:

```python
df = df_bikes

bike_column = ['Fahrradbruecke total']

# Defining the steep drop
steep_drop_before = ((df[bike_column].diff().abs() > 50) & (df[bike_column] == 0)).any(axis=1)
steep_drop_after = ((df[bike_column].diff(-1).abs() > 50) & (df[bike_column] == 0)).any(axis=1)

# Defining the sequences of at least 3 consecutive zeros
consecutive_zeros = df[bike_column].rolling(window=3).apply(lambda x: np.all(x == 0), raw=True).all(axis=1)

# row without any date
all_na = df[bike_column].isna().all(axis=1)

# Combining the two conditions
df['faulty'] = steep_drop_before | steep_drop_after | consecutive_zeros | all_na

# Find the start and end times of the faulty periods
faulty_periods = df.loc[df['faulty'] != df['faulty'].shift().fillna(False), ['Start DateTime', 'faulty']]

# Create dataframes for start and end times separately
start_times = faulty_periods.loc[faulty_periods['faulty'], 'Start DateTime']
end_times = faulty_periods.loc[~faulty_periods['faulty'], 'Start DateTime']

# If the last period is faulty and has no end, add the last timestamp of df as its end
if len(start_times) > len(end_times):
    end_times.loc[len(end_times)] = df['Start DateTime'].iloc[-1]

# Print start and end times
for start, end in zip(start_times, end_times):
    print(f"Faulty period started at {start} and ended at {end}")

```

More then one faulty timeframe have been detected. The start time is included, the end time excluded.

### TPZ Plots

```python
plt.figure(figsize=(14,6))

# Convert 'Start DateTime' to datetime if it's not already
df_TPZ['Start DateTime'] = pd.to_datetime(df_TPZ['Start DateTime'])

# Set 'Start DateTime' as the index
df = df_TPZ.set_index('Start DateTime')

# Sort data by 'Start DateTime' before plotting
df_sorted = df.sort_values('Start DateTime')

# Select only the columns to resample
df_to_resample = df_sorted[['TPZCH', 'TPZD']]

# Resample and aggregate by day
df_daily = df_to_resample.resample('D').sum()

plt.plot(df_daily.index, df_daily['TPZCH'], label='TPZCH', linestyle='solid')
plt.plot(df_daily.index, df_daily['TPZD'], label='TPZD', linestyle='solid')

plt.xlabel('Start DateTime')
plt.ylabel('Count')
plt.legend()
plt.title('Time Series Plot (Daily Aggregate)')
plt.show()

plt.figure(figsize=(14,6))

# Filter data for the specific day
df_day = df[df.index.date == pd.to_datetime('2021-01-15').date()]

plt.plot(df_day.index, df_day['TPZCH'], label='TPZCH', linestyle='solid')
plt.plot(df_day.index, df_day['TPZD'], label='TPZD', linestyle='solid')

plt.xlabel('Time')
plt.ylabel('Count')
plt.legend()
plt.title('Time Series Plot for 2021-01-15')
plt.show()
```
